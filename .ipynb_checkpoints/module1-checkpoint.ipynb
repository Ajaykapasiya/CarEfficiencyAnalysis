### Import necessary libraries


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt   
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    mean_squared_error, 
    mean_absolute_error, 
    r2_score
)
```

### Load the dataset

We will use the car dataset to analyze various factors influencing the car's miles per gallon (mpg).


```python
file_path = './Datasets/auto-mpg.csv'
df = pd.read_csv(file_path)
df.head() 
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.boxplot(x=df['horsepower'])
```

### Replace null value in Horsepower with median

```python
median=df['horsepower'].median()
df['horsepower']=df['horsepower'].fillna(median)
```

```python
df.isnull().sum()
```

### Pairplots to study correlation

```python
sns.pairplot(df[["mpg","cylinders","displacement","horsepower","weight","acceleration"]])
```

## Converting units to Indian standard

### cu inches to cu cm (cc)

```python
df['displacement_cc']=df['displacement']*16.387
```

### lbs to kg

```python
df['weight_kg']=df['weight']/2.205
```

### MPG to kpl(km per litre)

```python
df['mileage_kpl']=df['mpg']/2.352
```

After conversion

```python
df.head()
```

```python
df.info()
```

```python
df['horsepower']=df['horsepower'].astype(int)
```

### Mileage Vs Cylinders

```python
palette = sns.color_palette("Set2", len(df['cylinders'].unique()))  # Choose a color palette
plt.figure(figsize=(8,6))
sns.boxplot(y='mileage_kpl', x='cylinders', data=df, hue='cylinders', palette=palette, legend=False)
plt.xlabel("No. of cylinder")
```

### Pie-chart for Cylinder columns

```python
plt.figure(figsize=(14,8))
df.cylinders.value_counts().plot(kind='pie')
```

From this pie-chart it is clear that more than half vehicles have 4 cylinders
while other half have 6 or 8 cylinders with very few having other 

### Boxplots of performance over the years

```python
palette = sns.color_palette("Set2", len(df['cylinders'].unique())) 
plt.figure(figsize=(6, 4))
sns.boxplot(y='mileage_kpl', x='model-year', data=df, hue='cylinders', palette=palette, legend=False)

plt.xlabel("Model Year")          
plt.ylabel("Mileage (KPL)")     
plt.title("Mileage by Model Year and Number of Cylinders") 
plt.show()
```

### Mileage By Horsepower 

```python
plt.figure(figsize=(6,4))
plt.scatter(df.horsepower,df.mileage_kpl)
plt.xlabel("Horsepower")          
plt.ylabel("Mileage (KPL)")     
plt.title("Mileage by Horsepower") 
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df.weight_kg,df.mileage_kpl)
plt.xlabel("Weight")          
plt.ylabel("Mileage (KPL)")     
plt.title("Mileage by Weight") 
plt.show()
```

```python
df.head()
```

### Removing irrelevent columns

```python
X=df.drop(['displacement','weight','mpg','weight'],axis=1)
```

```python
X.head()
```

```python
# correlations
X.corr()
```

### Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(X.corr())
```

```python
print(X.corr()["mileage_kpl"].sort_values(ascending=False))
```

```python
y=X['mileage_kpl']
```

```python
X=X.drop(['mileage_kpl'],axis=1)
```

```python
X.head()
```

### Getting test and train sets

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
lrModel=LinearRegression()
```

```python
lrModel.fit(X_train,y_train)
```

```python
y_predicted=lrModel.predict(X_test)
```

```python
lrModel.score(X_test,y_test)
```

Evaluation Metrices:

```python
print("Mean absolute error: ", mean_absolute_error(y_test, y_predicted))
print("Mean squared error: ", mean_squared_error(y_test, y_predicted))
print("Root Mean squared error: ", np.sqrt(mean_squared_error(y_test, y_predicted)))
print("R2 score: ", r2_score(y_test, y_predicted))
```

### Predicting Mileage in kpl for given values

```python
lrModel.predict([[4,120,9.7,90,1479,1160]])
```

Hence the predicted output of the above car is :15.41 Km/l

# Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_rf=RandomForestRegressor(n_estimators=50)
```

```python
model_rf.fit(X_train,y_train)
```

```python
y_pred_rf=model_rf.predict(X_test)
```

```python
print("Mean absolute error: ", mean_absolute_error(y_test, y_pred_rf))
print("Mean squared error: ", mean_squared_error(y_test, y_pred_rf))
print("Root Mean squared error: ", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
print("R2 score: ", r2_score(y_test, y_pred_rf))
```
